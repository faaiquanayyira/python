```python
#oddnumber

f=input("Enter source file name:")
f1=input("Enter copy file name:")

with open(f,"r") as src, open(f1,"w") as dest:
    lines=src.readlines()
    for i in range(len(lines)):
        if i%2==0:
            dest.write(lines[i])
            
print("Odd lines copied successfully!")
```

```python
#read columns

import csv
f=input("Enter CSV file name:")
with open(f,"r")as f1:
    reader=csv.DictReader(f1)
    print("\nstudent details:")
    print("---------------")

    for row in reader:
        print("StudenName",row["Name"],"\tStudentAge:",row["Age"])
```

```python
#dict to csv

data=[
   {"Name":"Reshmi","Age":22,"Marks":85},
   {"Name":"Arun","Age":21,"Marks":78},
   {"Name":"Meera","Age":23,"Marks":91}
]

with open("Students_dict.csv","w",newline="")as file:
    fieldnames=["Name","Age","Marks"]
    writer=csv.DictWriter(file,fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data)

print("Dictionary data written successfully!")


with open("Students_dict.csv","r") as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```
